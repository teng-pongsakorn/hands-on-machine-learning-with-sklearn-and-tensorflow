```python
import numpy as np
import tensorflow as tf
import datetime
```

```python
def reset_graph(seed=42):
    tf.reset_default_graph()
    tf.set_random_seed(seed)
    np.random.seed(seed)
    
def make_logdir(prefix=''):
    base = 'tf_logs/'
    now = datetime.datetime.utcnow().strftime('%Y%m%d%H%M%S')
    if prefix:
        prefix += '-run-' + now + '/'
    else:
        prefix = 'run-' + now + '/'
    return base + prefix
```

#### Using the Estimator API (formerly tf.contrib.learn)


```python
(X_train, y_train), (X_test, y_test) = tf.keras.datasets.mnist.load_data()
```

```python
X_train = X_train.reshape((len(X_train), -1)).astype('float32') / 255.
X_test = X_test.reshape((len(X_test), -1)).astype('float32') / 255.
y_train = y_train.astype('int32')
y_test = y_test.astype('int32')

X_val, X_train = X_train[:5000], X_train[5000:]
y_val, y_train = y_train[:5000], y_train[5000:]
```

```python
X_train.shape, X_val.shape, X_test.shape
```

```python
y_train.shape, y_val.shape, y_test.shape
```

```python
feature_cols = [tf.feature_column.numeric_column(key='X', shape=(28*28))]
```

```python
dnn_clf = tf.estimator.DNNClassifier(hidden_units=[300, 100], feature_columns=feature_cols, n_classes=10)
```

```python
input_fn = tf.estimator.inputs.numpy_input_fn(x={'X':X_train}, y=y_train, 
                                              batch_size=50, shuffle=True, num_epochs=40)
```

```python
dnn_clf.train(input_fn=input_fn)
```

```python
test_input_fn = tf.estimator.inputs.numpy_input_fn(x={'X':X_test}, y=y_test, shuffle=False)
eval_result = dnn_clf.evaluate(input_fn=test_input_fn)
```

#### Using Plain TensorFlow

```python
n_inputs = 28*28
n_hidden1 = 300
n_hidden2 = 100
n_outputs = 10
```

```python
reset_graph()

my_logdir = make_logdir('my_dnn')
```

```python
print('logdir:', my_logdir)
```

```python
reset_graph()
X = tf.placeholder(dtype=tf.float32, shape=(None, n_inputs), name='X')
y = tf.placeholder(dtype=tf.int32, shape=(None), name='y')


def neural_layer(X, n_neurons, name, activation=None):
    with tf.name_scope(name):
        n_inputs = int(X.get_shape()[1])
        
        # init std = 2/sqrt(n)
        init = tf.truncated_normal(shape=(n_inputs, n_neurons), stddev=1./np.sqrt(n_inputs))
        
        W = tf.Variable(init, name='weights')
        b = tf.Variable(tf.zeros([n_neurons]), name='biases')
        z = tf.add(tf.matmul(X, W), b)
        if activation=='relu':
            return tf.nn.relu(z)
        return z
    
# model
with tf.name_scope('dnn'):
    hidden1 = neural_layer(X, n_hidden1, name='hidden1', activation='relu')
    hidden2 = neural_layer(hidden1, n_hidden2, name='hidden2', activation='relu')
    logits = neural_layer(hidden2, n_outputs, name='logits', activation=None)
    
# loss
with tf.name_scope('loss'):
    xentropy = tf.losses.sparse_softmax_cross_entropy(labels=y, logits=logits)
    loss = tf.reduce_mean(xentropy, name='loss')
    
# train
learning_rate = .01
with tf.name_scope('train'):
    optimizer = tf.train.AdamOptimizer(learning_rate)
    training_op = optimizer.minimize(loss)
    
# eval
with tf.name_scope('eval'):
    correct = tf.nn.in_top_k(logits, y, k=1)
    accuracy = tf.reduce_mean(tf.cast(correct, dtype='float32'), name='accuracy')
    
# init
with tf.name_scope('init'):
    init = tf.global_variables_initializer()
    
# save
with tf.name_scope('save'):
    saver = tf.train.Saver()
    
# log
with tf.name_scope('log'):
    loss_summary = tf.summary.scalar('xentropy_loss', loss)
    file_writer = tf.summary.FileWriter(my_logdir, graph=tf.get_default_graph())
```

```python
def shuffle_batch(X, y, batch_size):
    assert len(X) == len(y)
    rand_idx = np.random.permutation(len(X))
    for idx in np.array_split(rand_idx, batch_size):
        yield X[idx], y[idx]
```

```python
n_epochs = 400
batch_size = 50

with tf.Session() as sess:
    sess.run(init)
    
    for epoch in range(n_epochs):
        for x_batch, y_batch in shuffle_batch(X_train, y_train, batch_size):
            
            sess.run(training_op, feed_dict={X:x_batch, y:y_batch})
        loss_str = loss_summary.eval(feed_dict={X:x_batch, y:y_batch})
        file_writer.add_summary(loss_str)
        if epoch % 10 == 0:
            acc_train = accuracy.eval(feed_dict={X:x_batch, y:y_batch})
            acc_test = accuracy.eval(feed_dict={X:X_test, y:y_test})
            print('Epoch {:3d}, train acc: {:.4f}, test acc: {:.4f}'.format(epoch, acc_train, acc_test))
    
    save_path = saver.save(sess, save_path='./my_dnn_final.ckpt')
```

#### ex.9: train MNIST to achieve 98% accuracy

```python
X_train.shape
```

```python
n_input = 28*28
n_hidden1 = 300
n_hidden2 = 100
n_output = 10

learning_rate = .01
```

```python
logdir = make_logdir('dnn300-100-adam')
logdir
```

```python
reset_graph()

X = tf.placeholder(shape=(None, n_input), dtype=tf.float32, name='X')
y = tf.placeholder(shape=(None,), dtype=tf.int32, name='y')

with tf.name_scope('dnn'):
    hidden1 = tf.layers.dense(X, units=n_hidden1, activation=tf.nn.relu, name='hidden1')
    hidden2 = tf.layers.dense(hidden1, units=n_hidden2, activation=tf.nn.relu, name='hidden2')
    logits = tf.layers.dense(hidden2, units=n_output, activation=None, name='outputs')
    
with tf.name_scope('loss'):
    xentropy = tf.nn.sparse_softmax_cross_entropy_with_logits(labels=y, logits=logits)
    loss = tf.reduce_mean(xentropy, name='loss')
    tr_loss_summary = tf.summary.scalar(name='tr-loss', tensor=loss)
    val_loss_summary = tf.summary.scalar(name='val-loss', tensor=loss)
    
with tf.name_scope('train'):
    optimizer = tf.train.AdamOptimizer(learning_rate)
    training_op = optimizer.minimize(loss)
    
with tf.name_scope('eval'):
    y_pred = tf.argmax(logits, axis=1, name='y_pred', output_type=tf.int32)
    accuracy = tf.reduce_mean(tf.cast(tf.equal(y_pred, y), dtype=tf.float32), name='accuracy')
    tr_acc = tf.summary.scalar(name='tr-acc', tensor=accuracy)
    val_acc = tf.summary.scalar(name='val-acc', tensor=accuracy)
    
init = tf.global_variables_initializer()
saver = tf.train.Saver()
file_writer = tf.summary.FileWriter(logdir, graph=tf.get_default_graph())
```

```python
n_epochs = 100
batch_size = 50

with tf.Session() as sess:
    sess.run(init)
    
    for epoch in range(n_epochs):
        for x_batch, y_batch in shuffle_batch(X_train, y_train, batch_size):
            sess.run(training_op, feed_dict={X:x_batch, y:y_batch})
        
        print('epoch {:3d}, train-acc: {:.4f}, val-acc: {:.4f}'.format(epoch,
                                                                       sess.run(accuracy, 
                                                                                feed_dict={X:x_batch, y:y_batch}),
                                                                       sess.run(accuracy,
                                                                               feed_dict={X:X_val, y:y_val})))
        tr_acc_str = tr_acc.eval(feed_dict={X:x_batch, y:y_batch})
        val_acc_str = val_acc.eval(feed_dict={X:X_val, y:y_val})
        file_writer.add_summary(tr_acc_str, global_step=epoch)
        file_writer.add_summary(val_acc_str, global_step=epoch)
        
    y_test_pred = y_pred.eval(feed_dict={X:X_test})
```
